```python
type(df.loc[0,'cast'])
```

### as seen above cast column have list in str type but we need it in list object , so lates change it's data type to list

```python
def convert_to_list(str):
    return eval(str)                         #....eval fun evaluate str into python statement
```

```python
# applying this fun only to non null rows of cast column
df.loc[~df['cast'].isna(),'cast'] = df.loc[~df['cast'].isna(),'cast'].apply(convert_to_list)
```

```python
# same for crew column
df.loc[~df['crew'].isna(),'crew'] = df.loc[~df['crew'].isna(),'crew'].apply(convert_to_list)
```
